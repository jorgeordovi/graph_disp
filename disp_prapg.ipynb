```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + 1 + np.random.normal(0, 0.1, 50)
```

```python
plt.scatter(x, y, color='orange')
plt.title('Graph in Dispersion Mode Basic ')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
corr = np.corrcoef(x, y)[0, 1]
plt.scatter(x, y, color='blue')
plt.title(f'Graph Disprsion (Correlation: {corr:.2f})')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.show()
```

```python
plt.scatter(x, y, color='orange', label='Data')
m, b = np.polyfit(x,y,1)
plt.plot(x, m*x+b, color='white', label='Trend Line') 
plt.title(f'Dispersion Graph with Trend (Correlation: {corr:.2f})')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, color='red', marker='^', label='Data')
plt.title(f'Dispersion Graph with Personalized Caps')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.legend()
plt.grid(True, color='white')
plt.show()
```

```python
plt.scatter(x, y, color='orange', label='Data')
plt.title(f'Dispersion Graph with Personalized Caps')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.legend()
plt.grid(True, color='white')
plt.show()
```

```python
sizes = 1000 * np.random.rand(50)
colors= np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis', label='Data')
plt.title(f'Dispersion Bubble Graph Personalized')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.colorbar() # Adds color bar to hue
plt.grid(True, color='white')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

z = np.random.rand(50)

fig = plt.figure()
ax= fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='orange', marker='o')

ax.set_title('Advanced Dispersion Graph')
ax.set_xlabel('X Variable')
ax.set_ylabel('Y Variablre')
ax.set_zlabel('z Varible')

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(12, 6))

axs[0].scatter(x, y, color='red')
axs[0].set_title('Dispersion Graps 2')
axs[0].set_xlabel('X Variable')
axs[0].set_ylabel('Y Variable')

axs[1].scatter(y,x, color='blue')
axs[1].set_title('Dispersion Graph 2')
axs[1].set_xlabel('X Variable')
axs[1].set_ylabel('Y Variable')

plt.tight_layout()
plt.show()
```

```python
sizes = 1000 * np.random.rand(50)
colors= np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='plasma', label='Data')
plt.title(f'Dispersion Bubble Graph Personalized 2')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.colorbar(label='Intensity') # Adds color bar to hue
plt.grid(True, color='gray')
plt.show()
```
